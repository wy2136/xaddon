```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format ='retina'
plt.rcParams['figure.dpi'] = 150

import xaddon
```

## raw data 

```python
np.random.seed(0)
da = xr.DataArray(np.random.randn(100), dims='day', name='Ta', attrs={'units': 'degC'})
da.plot()
```

## power spectrum 

```python
da.go.power_spectrum().plot()
```

## find peaks 

```python
da.go.power_spectrum().plot()
da.go.power_spectrum() \
    .go.find_peaks().peaks.plot(marker='x', linestyle='none')
plt.savefig('power_spectrum_peaks.png')
```
